```python
import yt_dlp
import requests
import json
```

```python
video_url = "https://www.youtube.com/watch?v=qrXHjlNwxjs"
```

```python
ydl_opts = {
    'writesubtitles': True,
    'subtitleslangs': ['it'],  # Cambia con la lingua desiderata (es. 'it' per italiano)
    'skip_download': True,
    'quiet': True,
    'format': 'bestaudio/best',
    'outtmpl': '%(title)s.%(ext)s',
}
```

```python
with yt_dlp.YoutubeDL(ydl_opts) as ydl:
    info = ydl.extract_info(video_url, download=False)
    subtitles = info.get('automatic_captions', {}).get('it', [])  # Sostituisci con la lingua desiderata

    if subtitles:
        subtitle_url = subtitles[0]['url']
        print(f"Scarica i sottotitoli da: {subtitle_url}")
    else:
        print("Sottotitoli non disponibili.")
        
```

```python
def extract_subtitles(url):
    try:
        # Scarica il file JSON
        response = requests.get(url)
        response.raise_for_status()
        
        # Converte il testo in formato JSON
        data = json.loads(response.text)
        
        # Estrai i sottotitoli
        subtitles = []
        for event in data.get("events", []):
            for segment in event.get("segs", []):
                subtitles.append(segment.get("utf8", ""))
        
        # Unisci e restituisci i sottotitoli come testo
        return " ".join(subtitles)
    except requests.exceptions.RequestException as e:
        return f"Errore durante il download: {e}"
    except json.JSONDecodeError:
        return "Errore nella decodifica del JSON"

# Esempio di utilizzo
url = subtitle_url
subtitles_text = extract_subtitles(url)
#print(subtitles_text)
```
